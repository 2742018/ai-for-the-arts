# Designing a Creative Large Language Model Application

## Part 4: My Creative Design 

### What is a Large Language Model?
A large language model (LLM) is a type of artificial intelligence designed to understand
and generate human-like text based on the data it has been trained on.

## Creative Design – “PoetMate”

**PoetMate: A Creative Writing Assistant for Historical Poetic Styles**

## Introduction
**PoetMate** is a writing assistant powered by a large language model (LLM), designed to help users experiment with poetic language from across literary history. By blending creative prompting with stylistic emulation, the tool enables users to co-create original poems in the voice of different periods, authors, or traditions.

The goal is not to generate perfect poetry, but to support experimentation and learning in a low-pressure, interactive space. PoetMate is especially aimed at students, educators, and hobbyist writers across the arts and humanities.

## Interaction Design

### How users interact:
- Users select a poetic **style or influence** (e.g. Romanticism, Haiku, Harlem Renaissance).
- They enter a **theme or emotional prompt** (e.g. "grief", "revolution", "the night sky").
- The model generates an initial draft poem.
- Users can then ask:
  - “Add more metaphor”
  - “Explain the rhyme scheme”
  - “Can you make it sound more like Emily Dickinson?”
  - “Add a modern twist to the last line”

### What the AI does:
- Responds with stylistically consistent rewrites or elaborations
- Explains structural or linguistic choices
- Suggests historical context or related poets
- Highlights poetic devices used in the draft (e.g. enjambment, assonance)

### Tools involved:
- A chat-based web interface or notebook
- A fine-tuned GPT-like model with prompt-engineering for poetry modes
- (Optional) Access to a curated dataset of poetic examples
- (Optional) Export to PDF, writing portfolio, or zine format

## Value of the Design

### Learning
- Encourages deeper understanding of poetic structure and form
- Allows users to explore unfamiliar literary traditions through practice
- Supports humanities pedagogy through creative inquiry

### Engagement
- Makes poetry writing interactive and playful
- Promotes iteration and feedback rather than fear of failure
- Ideal for workshops, online courses, and informal learning

### Inclusion
- Supports users with different levels of literary experience
- Can accommodate multiple dialects, styles, and poetic heritages
- Makes poetic expression accessible for second-language learners and neurodiverse users

## Critical Reflection

PoetMate reflects both the potential and the challenges of using AI in creative arts contexts.

**Positive impacts:**
- It can democratise access to poetic techniques and histories.
- It supports learners who might otherwise feel intimidated by formal poetry.
- It builds bridges between literature and technology in an approachable way.

**Potential limitations and risks:**
- The model might reinforce dominant styles, neglecting less-represented voices.
- It may blur lines between human-authored and machine-generated work, raising authorship questions.
- Users might become over-reliant on suggestions, inhibiting their personal voice.

To address these concerns, the design includes reflective features such as:
- "Explain your choices" prompts
- Citations of literary inspirations
- A collaborative “edit mode” encouraging user revision

In summary, **PoetMate** is not just a tool to generate poems — it’s a creative scaffold that invites users to explore, adapt, and shape poetic language as part of a broader learning journey in the arts and humanities.
